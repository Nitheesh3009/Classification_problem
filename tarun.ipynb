```python
import numpy as np
import pandas as pd
from textblob import Blobber
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer
tb = Blobber(analyzer=NaiveBayesAnalyzer())

#print tb("sentence you want to test")
```

```python
df=pd.read_csv(r'D:\Assignments\Tarun\comments.csv')
df
```

```python
test2=[]
for i in df: 
    test2=(df[i])
print(len(test2))    
```

```python
len(test2)
```

```python
analysisp=[]
analysiss=[]
percent=[]
for i in test2:
    analysis=TextBlob(i)
    #df.insert(2,"polarity",analysis)
    x=(analysis.sentiment.polarity)*100
    percent.append(float("{0:.4f}".format(x)))
    if analysis.sentiment.polarity>0:
        analysisp.append('positive')
    else:
        analysisp.append('negative')
    print("polarity: "+str(analysis.sentiment.polarity))
    analysiss.append(analysis.sentiment.subjectivity)
    print("subjectivity: "+str(analysis.sentiment.subjectivity))
    
```
